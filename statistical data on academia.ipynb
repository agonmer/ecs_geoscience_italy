```python
# Plot for the paper Bazzicalupo et al.

idM_as = (df_as.GENERE=='M')
idF_as = (df_as.GENERE=='F')

idgrado_ps = (df_as.GRADE == 'D'); str_grade = 'post-doctoral scholars'; ylim = [-50,3200]
idgrado_r = (df_as.GRADE == 'C'); str_grade = 'researchers'; ylim = [800,7200]
idgrado_ap = (df_as.GRADE == 'B'); str_grade = 'associate professors'; ylim = [500,6000]
idgrado_fp = (df_as.GRADE == 'A'); str_grade = 'full professors'; ylim = [-200,8500]

id_ate = (df_as.CODICE_ATENEO == 'TTTTT')

id_25 = (df_as.CLASSE_ETA=='< 35')
id_35 = (df_as.CLASSE_ETA=='35 - 44')
id_45 = (df_as.CLASSE_ETA=='45 - 54')
id_55 = (df_as.CLASSE_ETA=='>= 55')

tempo = df_as.ANNO[idgrado_ps][id_ate][idM_as][id_35]

# Post-docs
ps_25_m = df_as.N_AcStaff[idgrado_ps][id_ate][idM_as][id_25].values
ps_25_f = df_as.N_AcStaff[idgrado_ps][id_ate][idF_as][id_25].values
ps_35_m = df_as.N_AcStaff[idgrado_ps][id_ate][idM_as][id_35].values
ps_35_f = df_as.N_AcStaff[idgrado_ps][id_ate][idF_as][id_35].values
ps_45_m = df_as.N_AcStaff[idgrado_ps][id_ate][idM_as][id_45].values
ps_45_f = df_as.N_AcStaff[idgrado_ps][id_ate][idF_as][id_45].values
ps_55_m = df_as.N_AcStaff[idgrado_ps][id_ate][idM_as][id_55].values
ps_55_f = df_as.N_AcStaff[idgrado_ps][id_ate][idF_as][id_55].values

# Researchers
r_25_m = df_as.N_AcStaff[idgrado_r][id_ate][idM_as][id_25].values
r_25_f = df_as.N_AcStaff[idgrado_r][id_ate][idF_as][id_25].values
r_35_m = df_as.N_AcStaff[idgrado_r][id_ate][idM_as][id_35].values
r_35_f = df_as.N_AcStaff[idgrado_r][id_ate][idF_as][id_35].values
r_45_m = df_as.N_AcStaff[idgrado_r][id_ate][idM_as][id_45].values
r_45_f = df_as.N_AcStaff[idgrado_r][id_ate][idF_as][id_45].values
r_55_m = df_as.N_AcStaff[idgrado_r][id_ate][idM_as][id_55].values
r_55_f = df_as.N_AcStaff[idgrado_r][id_ate][idF_as][id_55].values

# Associate professors
ap_25_m = df_as.N_AcStaff[idgrado_ap][id_ate][idM_as][id_25].values
ap_25_f = df_as.N_AcStaff[idgrado_ap][id_ate][idF_as][id_25].values
ap_35_m = df_as.N_AcStaff[idgrado_ap][id_ate][idM_as][id_35].values
ap_35_f = df_as.N_AcStaff[idgrado_ap][id_ate][idF_as][id_35].values
ap_45_m = df_as.N_AcStaff[idgrado_ap][id_ate][idM_as][id_45].values
ap_45_f = df_as.N_AcStaff[idgrado_ap][id_ate][idF_as][id_45].values
ap_55_m = df_as.N_AcStaff[idgrado_ap][id_ate][idM_as][id_55].values
ap_55_f = df_as.N_AcStaff[idgrado_ap][id_ate][idF_as][id_55].values

# Full professors
fp_25_m = df_as.N_AcStaff[idgrado_fp][id_ate][idM_as][id_25].values
fp_25_f = df_as.N_AcStaff[idgrado_fp][id_ate][idF_as][id_25].values
fp_35_m = df_as.N_AcStaff[idgrado_fp][id_ate][idM_as][id_35].values
fp_35_f = df_as.N_AcStaff[idgrado_fp][id_ate][idF_as][id_35].values
fp_45_m = df_as.N_AcStaff[idgrado_fp][id_ate][idM_as][id_45].values
fp_45_f = df_as.N_AcStaff[idgrado_fp][id_ate][idF_as][id_45].values
fp_55_m = df_as.N_AcStaff[idgrado_fp][id_ate][idM_as][id_55].values
fp_55_f = df_as.N_AcStaff[idgrado_fp][id_ate][idF_as][id_55].values

fig = plt.figure(figsize=(14,8))
ax1 = plt.subplot(221)
ax2 = plt.subplot(222, sharex=ax1)
ax3 = plt.subplot(223, sharex=ax1)
ax4 = plt.subplot(224, sharex=ax1)

phd_m = np.flip(np.flip(df_phd_year.Isc_M.values)[idtime])
phd_f = np.flip(np.flip(df_phd_year.Isc_F.values)[idtime])
#ax1.plot(ay[idtime],np.flip(df_phd_year.Isc_M.values)[idtime],'-^')
#ax1.plot(ay[idtime],np.flip(df_phd_year.Isc_F.values)[idtime],'-vr')
ax1.plot(tempo,phd_m,'-^')
ax1.plot(tempo,phd_f,'-vr')
ax1.tick_params(axis='both',labelsize=14)
ax1.grid()
ax1.set_title('(A) PhD students',fontsize=14)
ax1.legend(['Males','Females'],fontsize=14)

ax2.plot(tempo,ps_25_m+ps_35_m+ps_45_m+ps_55_m,'-^')
ax2.plot(tempo,ps_25_f+ps_35_f+ps_45_f+ps_55_f,'-vr')
ax2.tick_params(axis='both',labelsize=14)
ax2.grid()
ax2.set_title('(B) Post-doctoral scholars',fontsize=14)

ax3.plot(tempo,r_25_m+r_35_m+r_45_m+r_55_m,'-^')
ax3.plot(tempo,r_25_f+r_35_f+r_45_f+r_55_f,'-vr')
ax3.tick_params(axis='both',labelsize=14)
ax3.grid()
ax3.set_title('(C) Researchers',fontsize=14)
ax3.set_xlabel('Year',fontsize=14)

ax4.plot(tempo,ap_25_m+ap_35_m+ap_45_m+ap_55_m,'-^')
ax4.plot(tempo,ap_25_f+ap_35_f+ap_45_f+ap_55_f,'-vr')
ax4.plot(tempo,fp_35_m+fp_45_m+fp_55_m,'--^',color='C0')
ax4.plot(tempo,fp_35_f+fp_45_f+fp_55_f,'--vr')
ax4.plot(tempo,np.zeros_like(tempo),'-k',label='A.P.')
ax4.plot(tempo,np.zeros_like(tempo),'--k',label='F.P.')
ax4.set_ylim(2000,14500)
ax4.tick_params(axis='both',labelsize=14)
ax4.grid()
ax4.set_title('(D) Professors',fontsize=14)
ax4.set_xlabel('Year',fontsize=14)
ax4.legend(fontsize=14)

figure_name = 'numbers_categories_male_female.png'
plt.savefig(figure_name,bbox_inches='tight',dpi=300)
```
